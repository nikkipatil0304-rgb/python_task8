```python
import re

def is_palindrome(text):
    # Clean the string: remove non-alphanumeric characters and lowercase it
    clean_text = re.sub(r'[^a-zA-Z0-9]', '', text).lower()
    # Check if the string is equal to its reverse
    return clean_text == clean_text[::-1]

sent1 = 'A man, a plan, a canal: Panama'

if is_palindrome(sent1):
    print(f"'{sent1}' is a palindrome!")
else:
    print(f"'{sent1}' is not a palindrome.")
```

```python
# 1. Login Function
def login_system():
    username = input("Enter username: ")
    password = input("Enter password: ")
    # Simple hardcoded check
    if username == "admin" and password == "1234":
        print("Login Successful!\n")
        return True
    else:
        print("Access Denied.")
        return False

# 2. Palindrome Check Function
def check_sentence(text):
    clean = "".join(char.lower() for char in text if char.isalnum())
    return clean == clean[::-1]

# 3. String Statistics Function
def get_stats(text):
    return f"Length: {len(text)} characters"

# 4. Main Controller Function
def run_app():
    if login_system():
        sent1 = 'A man, a plan, a canal: Panama'
        print(f"Processing: {sent1}")
        
        if check_sentence(sent1):
            print("Result: This is a palindrome.")
        
        print(get_stats(sent1))
        print("Application Finished.")

```
